```python
import pandas as pd
import numpy as np
import geopandas as gpd
import xarray as xr
import matplotlib.pyplot as plt 
import rasterio
from rasterio.windows import Window
import os
from src.config import *
```

```python
data_tif = '/home/ajai-krishna/work/GEO_AI/data/outputs/ground_points.tif'
output_dir = '/home/ajai-krishna/work/GEO_AI/data/outputs/segmented_data/'
```

```python
tile_size = 256

os.makedirs(output_dir, exist_ok=True)

with rasterio.open(data_tif) as src:
    
    width = src.width
    height = src.height
    meta = src.meta.copy()

    tile_id = 0

    for i in range(0, width, tile_size):
        for j in range(0, height, tile_size):

            window = Window(i, j, tile_size, tile_size)

            transform = src.window_transform(window)

            tile = src.read(window=window)

            meta.update({
                "height": tile.shape[1],
                "width": tile.shape[2],
                "transform": transform
            })

            output_file = f"{output_dir}/tile_{tile_id}.tif"

            with rasterio.open(output_file, "w", **meta) as dst:
                dst.write(tile)

            tile_id += 1

print("Tiles created:", tile_id)
```
